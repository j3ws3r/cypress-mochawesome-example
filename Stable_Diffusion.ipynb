```python
from PIL import Image

def image_grid(imgs, rows, cols):
    assert len(imgs) == rows*cols

    w, h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    grid_w, grid_h = grid.size
    
    for i, img in enumerate(imgs):
        grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid
```

# Inference Steps

```python
from pathlib import Path
resultPath = Path("./inferenceSteps/")

if resultPath.is_dir():
  print(f"{resultPath} directory already exists... skipping download")
else: 
  print(f"{resultPath} does not exist, creating one...")
  resultPath.mkdir(parents=True, exist_ok=True)
```

```python
prompt = "Realistic image of a panda playing the guitar"
inferenceSteps = list(range(1,50))

for steps in inferenceSteps:
  image = pipe(prompt, guidance_scale=7, num_inference_steps=steps).images[0]
  image.save(f'./inferenceSteps/inference{str(steps)}.jpg', 'JPEG')
```

```python
import shutil
shutil.make_archive('./inferenceSteps', format='zip', root_dir='./inferenceSteps/')
```
